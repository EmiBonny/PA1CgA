```python
#escrever um programa que classifica os triangulos
#verificar se forma um triangulo valido
#tipo:
#equilatero
#isosceles
#escaleno

print ("|------------------------------------------------------------|")
print ("|                                                            |")
print ("|-----------------Classificador de triãngulos----------------|")
print ("|------------------------------------------------------------|")
lado1 = int (input("Digite o primeiro lado: "))
lado2 = int (input("Digite o segundo lado: "))
lado3 = int (input("Digite o terceiro lado: "))

#processamento e saida
if lado1 == lado2 == lado3:
  print (f"O triangulo é equilatero")
elif lado1 == lado2 or lado2 == lado3 or lado1 == lado3:
  print (f"O triangulo é isosceles")
else: lado1 != lado2 != lado3
print (f" O triangulo é escaleno")


```
