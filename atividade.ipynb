# Questão 3- Verificação de triângulo

```python
lado1 = int (input("Digite a medida do primeiro lado: "))
lado2 = int (input("Digite a medida do segundo lado: "))
lado3 = int (input("Digite a medida do terceiro lado"))
c = lado1 + lado2
if lado3 <= c:
    print (f"Os lados constituem um triângulo.")
else:
    print (f"Os lados não formam um triângulo.")

```

Questão 4- Sistema de energia

```python
vida = int (input("Digite a quantidade de energia do personagem: "))
if vida == 0:
    print (f"O personagem morreu.")
elif vida >= 1 and vida <= 30:
    print (f"Cuidado! Estado crítico.")
elif vida >= 31 and vida <= 70:
    print (f"Vida em estado normal.")
elif vida >= 71 and vida <= 100:
    print (f"O personagem está bem.")
```

Questão 5- Média escolar

```python
nota1 = float (input("insira a primeira nota: "))
nota2 = float (input("insira a segunda nota: "))
nota3 = float (input("insira a terceira nota: "))
nota4 = float (input("insira a quarta nota: "))
media = (nota1 + nota2 + nota3 + nota4) / 4
if media < 5:
    print (f"Sua média é {media} você foi reprovado.")
elif media >= 5 and media <= 6.9:
    print (f"Sua média é {media} você está de recuperação.")
elif media >= 7:
    print (f"Sua média é {media} você está aprovado.")

```
