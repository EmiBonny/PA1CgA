Escreve um algoritmo e o programa que recebe 04 notas do aluno e calcula a média depois de mostrar a média

```python
#entrada
notaum = 5
notadois = 7
notatres = 10
notaquatro = 6
#processamento
adicao = notaum + notadois + notatres + notaquatro
media = adicao / 4
#saída
print (f"média do aluno: {media}")
```
