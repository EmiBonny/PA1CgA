```python
#escreva um programa que:
#peça o valor da compra
#se o valor for maior que 100, aplique um desconto de 10%
#se não mostre o valor normal

valor = float (input ("Digite o valor da compra: "))
if valor > 100:
  desconto = valor * 0.9
  print (desconto)
else:
  print (valor)
```

```python
#escrever um programa que o aluno digita 04 notas dos bimestres e o programa calcula a média e mostra na saída se o aluno foi aprovado, recuperação ou reprovado.
nota1 = float (input ("digite a nota do 1 bimestre: "))
nota2 = float (input ("digite a nota do 2 bimestre: "))
nota3 = float (input ("digite a nota do 3 bimestre: "))
nota4 = float (input ("digite a nota do 4 bimestre: "))
media = (nota1 + nota2 + nota3 + nota4)/4

if media >= 7:
  print (f"você foi aprovado, sua média foi: {media}")
elif media >= 5:
  print (f"você está de recuperação, sua média foi: {media}")
else:
  print (f"você foi reprovado, sua média foi: {media}")
```
