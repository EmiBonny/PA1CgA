```python

a = int (input ("Digite o primeiro número: "))
b = int (input ("Digite o segundo número: "))

soma = a + b
subtração = a - b
multiplicação = a * b
divisão = a / b
divisão_int = a // b
resto = a % b

print (f"  A soma de {a} mais {b} é igual a: {soma}")
print (f"  A subtração de {a} menos {b} é igual a: {subtração}")
print (f"  A multiplicação de {a} vezes {b} é igual a: {multiplicação}")
print (f"  A divisão de {a} por {b} é igual a: {divisão}")
print (f"  A divisão de {a} por {b} é igual a: {divisão_int}")
print (f"  A divisão de {a} por {b} é igual a: {resto}")
```

```python


N1 = int (input ("Digite o primeiro número: "))
N2 = int (input ("Digite o segundo número: "))

igual = N1 == N2
diferente = N1 != N2
maior = N1 > N2
menor = N1 < N2
menor_igual = N1 <= N2
maior_igual = N1 >= N2

print(f"{N1} é igual a {N2}?: {igual}")
print(f"{N1} é diferente de {N2}?: {diferente}")
print(f"{N1} é maior que {N2}?: {maior}")
print(f"{N1} é menor que {N2}?: {menor}")

print(f"{N1} é menor ou igual a {N2}?: {menor_igual}")

print(f"{N1} é maior ou igual a {N2}?: {maior_igual}")

```
